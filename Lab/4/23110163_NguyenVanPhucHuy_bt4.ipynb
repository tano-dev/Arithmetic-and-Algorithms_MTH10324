> 1. Giải hệ 2 phương trình đồng dư tuyến tính 2 ẩn

```python
def ModInverse(a, m):
    m0, x0, x1 = m, 0, 1
    if m == 1:
        return 0
    while a > 1:
        q = a // m
        m, a = a % m, m
        x0, x1 = x1 - q * x0, x0
    if x1 < 0:
        x1 += m0
    return x1
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
def Solve2x2Congruence(a, b, c, d, e, f, m):
    Delta = (a * d - b * c) % m
    dDelta = gcd(Delta, m)
    if dDelta == 1:
        Delta_inv = ModInverse(Delta, m)
        x = (Delta_inv * (e * d - b * f)) % m
        y = (Delta_inv * (a * f - c * e)) % m
        return (x, y)
    else:
        return "Thất bại: Hệ không có nghiệm duy nhất"
    
    
# Ví dụ sử dụng
a, b, c, d, e, f, m = 2, 3, 1, 4, 5, 6, 7
x, y = Solve2x2Congruence(a, b, c, d, e, f, m)
print(f"Het phuong trinh: {a}x + {b}y ≡ {e} (mod {m}) va {c}x + {d}y ≡ {f} (mod {m})")
print(f"Nghiem (x, y) modulo {m}: ({x}, {y})")
```

> 2. Nghịch đảo của ma trận n × n modulo m

```python
def determinant(A, m):
    n = len(A)
    if n == 1:
        return A[0][0] % m
    det = 0
    for c in range(n):
        det += ((-1) ** c) * A[0][c] * determinant([row[:c] + row[c+1:] for row in A[1:]], m)
        det %= m
    return det
def adjugate(A, m):
    n = len(A)
    adj = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            Mji = determinant([row[:j] + row[j+1:] for row in A[:i] + A[i+1:]], m)
            adj[j][i] = ((-1) ** (i + j)) * Mji % m
    return adj
def MatrixInverseMod(A, m):
    D = determinant(A, m)
    if gcd(D, m) != 1:
        return "Thất bại: Ma trận không khả nghịch modulo m"
    D_inv = ModInverse(D, m)
    adj_A = adjugate(A, m)
    A_inv = [[(D_inv * adj_A[i][j]) % m for j in range(len(A))] for i in range(len(A))]
    return A_inv

# Ví dụ sử dụng
A = [[2, 3], [1, 4]]
m = 7
A_inv = MatrixInverseMod(A, m)
print("Ma trận nghịch đảo modulo 7:")
for row in A_inv:
    print(row)
    
    
```

> 3. Giải hệ n phương trình đồng dư n ẩn

```python
def SolveNxNCongruence(A, B, m):
    A_inv = MatrixInverseMod(A, m)
    if isinstance(A_inv, str):
        return "Hệ không có nghiệm duy nhất do ma trận hệ số không khả nghịch"
    n = len(A)
    X = [0] * n
    for i in range(n):
        for j in range(n):
            X[i] = (X[i] + A_inv[i][j] * B[j][0]) % m
    return X


# Ví dụ sử dụng
A = [[2, 3], [1, 4]]
B = [[5], [6]]
m = 7
X = SolveNxNCongruence(A, B, m)
print("Nghiem X:")
for x in X:
    print(x)
    
```

> 4. Kiểm tra giả nguyên tố Fermat cơ số b

```python
def FermatPseudoprimeTest(n, b):
    if gcd(b, n) > 1:
        return False
    r = pow(b, n - 1, n)
    return r == 1
# Ví dụ sử dụng
n = 561
b = 2
result = FermatPseudoprimeTest(n, b)
print(f"result Fermat cho n={n} và b={b}: {result}")
```

> 5. Kiểm tra Miller cơ số b (Số giả nguyên tố mạnh)

```python
def MillerTest(n, b):
    if n % 2 == 0 or b >= n:
        return False
    s, t = 0, n - 1
    while t % 2 == 0:
        t //= 2
        s += 1
    x = pow(b, t, n)
    if x == 1 or x == n - 1:
        return True
    for _ in range(s - 1):
        x = pow(x, 2, n)
        if x == n - 1:
            return True
    return False
# Ví dụ sử dụng
n = 561
b = 2
result = MillerTest(n, b)
print(f"result Miller cho n={n} và b={b}: {result}")
```


> 6. Kiểm tra tính nguyên tố tất định (n < 25 × 109
)

```python
def DeterministicPrimalityTest(n):
    if n <= 1:
        return "Hợp số"
    if n in {2, 3, 5, 7}:
        return "Số nguyên tố"
    if n % 2 == 0 or n % 3 == 0 or n % 5 == 0 or n % 7 == 0:
        return "Hợp số"
    B = [2, 3, 5, 7]
    for b in B:
        if not MillerTest(n, b):
            return "Hợp số"
    return "Số nguyên tố"


# Ví dụ sử dụng
n = 29
result = DeterministicPrimalityTest(n)
print(f"result kiểm tra tính nguyên tố tất định cho n={n}: {result}")

```
