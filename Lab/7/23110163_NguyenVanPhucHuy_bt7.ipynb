> 1. Viết chương trình tìm cấp của phần tử a theo modulo m (a,m) = 1

```python
# Cách 1: for i to m-1 do:
#   if a^i mod m = 1 then return i
def power(a, i, m):
    res = 1
    for _ in range(i):
        res = (res * a) % m
    return res
def order(a, m):
    for i in range(1, m):
        if power(a, i, m) == 1:
            return i
    return -1
a = int(input("Nhập a: "))
m = int(input("Nhập m: "))
if order(a, m) != -1:
    print(f"Cấp của {a} theo modulo {m} là: {order(a, m)}")
else:    print(f"{a} không có cấp theo modulo {m}")

# Cách 2 for i in {phi(m)} do:
#   if a^i mod m = 1 then return i

def phi(n):
    result = n
    p = 2
    while p * p <= n:
        if n % p == 0:
            while n % p == 0:
                n //= p
            result -= result // p
        p += 1
    if n > 1:
        result -= result // n
    return result
def order(a, m):
    phi_m = phi(m)
    for i in range(1, phi_m + 1):
        if power(a, i, m) == 1:
            return i
    return -1
a = int(input("Nhập a: "))
m = int(input("Nhập m: "))
if order(a, m) != -1:
    print(f"Cấp của {a} theo modulo {m} là: {order(a, m)}")
else:    
    print(f"{a} không có cấp theo modulo {m}")
```

> 2. Viết chương trình kiểm tra a có là căn nguyên thủy modulo p hay không?

```python
def is_primitive_root(a, p):
    if a <= 1 or a >= p:
        return False
    order_a = order(a, p)
    return order_a == phi(p)
a = int(input("Nhập a: "))
p = int(input("Nhập p: "))
if is_primitive_root(a, p):
    print(f"{a} là căn nguyên thủy modulo {p}")
else:
    print(f"{a} không phải là căn nguyên thủy modulo {p}")
```

> 3. Viết chương trình kiểm tra n có là căn nguyên thủy không? Nếu có thì tìm tất cả các căn nguyên thủy modulo n.

```python
def is_primitive_root(a, n):
    if a <= 1 or a >= n:
        return False
    order_a = order(a, n)
    return order_a == phi(n)
n = int(input("Nhập n: "))
def find_primitive_roots(n):
    primitive_roots = []
    for a in range(2, n):
        if is_primitive_root(a, n):
            primitive_roots.append(a)
    if not primitive_roots:
        print(f"Không có căn nguyên thủy modulo {n}")
    return primitive_roots
print(f"Các căn nguyên thủy modulo {n} là: {find_primitive_roots(n)}") 
```
