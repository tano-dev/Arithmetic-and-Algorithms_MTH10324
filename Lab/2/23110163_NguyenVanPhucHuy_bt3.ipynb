> 2.4. Giải Đồng Dư Tuyến Tính 

```python
def extended_euclid(a, b):
    if b == 0:
        return a, 1, 0
    g, x1, y1 = extended_euclid(b, a % b)
    x = y1
    y = x1 - (a // b) * y1
    return g, x, y

def solve(a, c, m):
    g, x0, _ = extended_euclid(a, m)
    if c % g != 0:
        return "Không có nghiệm"
    
    x1 = (x0 * (c // g)) % m
    step = m // g
    solutions = []
    
    for i in range(g):
        solution = (x1 + i * step) % m
        solutions.append(solution)
    
    return solutions

# Ví dụ: 3x ≡ 6 (mod 9)
a = 3
c = 6
m = 9
result = solve(a, c, m)
print("Nghiệm:", result)
# Kiểm tra: 3 ·2 = 6 ≡6, 3 ·5 = 15 ≡6, 3 ·8 = 24 ≡6 (mod 9)
for x in result:
    print(f"3 x {x} ≡ {(3 * x) % m} (mod {m})")
    


# 1. Find all solutions of each of the following linear congruences.
# a) 2x = 5 (mod 7) c)19x =  30 (mod 40)  e) 103x =  444 (mod 999)
# b)3x = 6 (mod 9) d)9x = 5 (mod 25) f)980x = 1500 (mod 1600)
# 2.  Find all solutions of each of the following linear congruences.
# a)3x = 2 (mod 7) c)17x = 14 (mod 21)  e) 128x = 833 (mod 1001)
# b)6x = 3 (mod 9) d)15x = 9 (mod 25) f)987x = 610 (mod 1597)


# 1a
a = 2
c = 5
m = 7
result = solve(a, c, m)
print("Nghiệm 1a:", result)
# 1b
a = 3
c = 6
m = 9
result = solve(a, c, m)
print("Nghiệm 1b:", result)
# 1c
a = 19
c = 30
m = 40
result = solve(a, c, m)
print("Nghiệm 1c:", result)
# 1d
a = 9
c = 5
m = 25
result = solve(a, c, m)
print("Nghiệm 1d:", result)
# 1e
a = 103
c = 444
m = 999
result = solve(a, c, m)
print("Nghiệm 1e:", result)
# 1f
a = 980
c = 1500
m = 1600
result = solve(a, c, m)
print("Nghiệm 1f:", result)


# 2a
a = 3
c = 2
m = 7
result = solve(a, c, m)
print("Nghiệm 2a:", result)
# 2b
a = 6
c = 3
m = 9
result = solve(a, c, m)
print("Nghiệm 2b:", result)
# 2c
a = 17
c = 14
m = 21
result = solve(a, c, m)
print("Nghiệm 2c:", result)
# 2d
a = 15
c = 9
m = 25
result = solve(a, c, m)
print("Nghiệm 2d:", result)
# 2e
a = 128
c = 833
m = 1001
result = solve(a, c, m)
print("Nghiệm 2e:", result)
# 2f
a = 987
c = 610
m = 1597
result = solve(a, c, m)
print("Nghiệm 2f:", result)

```

> 3.1. Phép thử chia

```python
def TrialDivision(n):
    factors = []
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors.append(d)
            n //= d
        d += 1
    if n > 1:
        factors.append(n)
    return factors
# Ví dụ: Phân tích n = 360
factors = TrialDivision(360)
print("Các nhân tử nguyên tố của 360:", factors)

factors = TrialDivision(23110163)
print("Các nhân tử nguyên tố của 23110163:", factors)

factors = TrialDivision(23110163 * 123456789)
print("Các nhân tử nguyên tố của 23110163 * 123456789:", factors)
```

> 3.2. Phương pháp Fermat

```python
def FermatFactorization(n):
    if n <= 1 or n % 2 == 0:
        return "n phải là số nguyên lẻ lớn hơn 1"
    
    x = int(n**0.5) + 1
    t = x**2 - n
    
    while int(t**0.5)**2 != t:
        x += 1
        t = x**2 - n
    
    y = int(t**0.5)
    
    if x - y > 1 and x + y < n:
        return (x - y, x + y)
    
    return "Thất bại hoặc n nguyên tố"

# Ví dụ: Phân tích n = 5959
factors = FermatFactorization(69421)
print("Các ước của 69421:", factors)
```

> 3.3. Phương pháp Pollard Rho

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a


def PollardRho(n, c):
    if n <= 1:
        return "n phải là số nguyên lớn hơn 1"
    if c == 0 or c == -2:
        return "c phải khác 0 và -2"
    
    def f(x):
        return (x**2 + c) % n
    
    x = 2
    y = 2
    d = 1
    
    while d == 1:
        x = f(x)  # Rùa: 1 bước
        y = f(f(y))  # Thỏ: 2 bước
        d = gcd(abs(y - x), n)
    
    if d != n:
        return d  # Ước không tầm thường
    else:
        return "Thất bại – thử giá trị c hoặc x0 khác"
    
# Ví dụ: Phân tích n = 69421 với c = 1
factor = PollardRho(69421, 1)
print("Một ước không tầm thường của 69421:", factor)


def PollardRhoBrent(n, c, m=128):
    if n <= 1:
        return "n phải là số nguyên lớn hơn 1"
    if c == 0 or c == -2:
        return "c phải khác 0 và -2"
    
    def f(x):
        return (x**2 + c) % n
    
    y = 2
    r = 1
    q = 1
    d = 1
    
    while d == 1:
        x = y
        for i in range(1, r + 1):
            y = f(y)
        
        k = 0
        while k < r and d == 1:
            ysave = y
            for i in range(1, min(m, r - k) + 1):
                y = f(y)
                q = (q * abs(x - y)) % n
            
            d = gcd(q, n)
            k += m
        
        r *= 2
    
    if d == n:
        d = 1
        y = ysave
        while d == 1:
            y = f(y)
            d = gcd(abs(x - y), n)
    
    return d

# Ví dụ: Phân tích n = 69421 với c = 1 và m = 128
factor = PollardRhoBrent(69421, 1, 128)
print("Một ước không tầm thường của 69421:", factor)
```

> 3.4. Phương pháp Pollard p − 1

```python
def PollardPMinus1(n, B):
    if n <= 1:
        return "n phải là số nguyên lớn hơn 1"
    
    a = 2
    primes = TrialDivision(B)  # Lấy các số nguyên tố ≤ B
    
    for p in primes:
        e = int(n**0.5) // p  # Lũy thừa tối đa của p trong M
        a = pow(a, pow(p, e), n)  # a ← a^(p^e) mod n
    
    d = gcd(a - 1, n)
    
    if 1 < d < n:
        return d
    elif d == 1:
        return "Thất bại: tăng B"
    else:
        return "Thất bại: d = n, thử a khác hoặc sang giai đoạn 2"
    
# Ví dụ: Phân tích n = 69421 với B = 10
factor = PollardPMinus1(69421, 10)
print("Một ước không tầm thường của 69421:", factor)
```
