```python
def convert10to2(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        n,r = div(n, 2)
        binary = str(r) + binary
        n //= 2
    return binary

def decimal_to_binary(n):
    if n == 0:
        return "0"
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary
def binary_to_decimal(b):
    decimal = 0
    for i in range(len(b)):
        decimal += int(b[-(i+1)]) * (2 ** i)
    return decimal
n = int(input("Nhập số thập phân: "))
binary = decimal_to_binary(n)
print(f"Số nhị phân: {binary}")
b = input("Nhập số nhị phân: ")
decimal = binary_to_decimal(b)
print(f"Số thập phân: {decimal}")
```

> 3. Viết chương trình chuyển một số từ cơ b1 sang cơ b2.

```python
def convert_base(num, b1, b2):
    # b1
    decimal = 0
    for i in range(len(num)):
        decimal += int(num[-(i+1)]) * (b1 ** i)
    
    # b2
    if decimal == 0:
        return "0"
    result = ""
    while decimal > 0:
        result = str(decimal % b2) + result
        decimal //= b2
    return result
num = input("Nhập số: ")
b1 = int(input("Nhập cơ b1: "))
b2 = int(input("Nhập cơ b2: "))
converted = convert_base(num, b1, b2)
print(f"Số sau khi chuyển từ cơ {b1} sang cơ {b2}: {converted}")
```

> 4. Given two positive integers m and n and their list of positive divisors. Find(m, n).

```python
def gcd(m, n):
    if m == 0: return n
    if n == 0: return m
    while m != n:
        if m > n:
            m = m - n
        else:
            n = n - m
    return m

m = int(input("Nhập số nguyên dương m: "))
n = int(input("Nhập số nguyên dương n: "))
result = gcd(m, n)
print(f"Ước chung lớn nhất của {m} và {n} là: {result}")
```

> 5. Given two integers, use the Euclidean algorithm to find their greatest common divisor.

```python
def euclidean_algorithm(a, b):
    while b:
        a, b = b, a % b
    return a
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
result = euclidean_algorithm(a, b)
print(f"Ước chung lớn nhất của {a} và {b} là: {result}")
```

> 6. Given two integers, find their greatest common divisor using the modified Euclidean algorithm.

```python
def euclidean_extended_algorithm(a, b):
    
    r0, r1 = a, b
    t0, t1 = 1, 0
    s0, s1 = 0, 1
    while r1 != 0:
        q = r0 // r1
        r0, r1 = r1, r0 - q * r1
        t0, t1 = t1, t0 - q * t1
        s0, s1 = s1, s0 - q * s1
    return r0, t0, s0

    
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
result = euclidean_extended_algorithm(a, b)
print(f"Ước chung lớn nhất của {a} và {b} là: {result}")
```

> Giải Phương Trình Diophantine

> Cho ba số nguyên a, b, c. Giải ax+by = c; tìm nghiệm nguyên (x, y) hoặc kết luận không
có nghiệm.

```python
def solve_diophantine(a, b, c):
    g, x0, y0 = euclidean_extended_algorithm(a, b)
    if c % g != 0:
        return "Phương trình không có nghiệm nguyên"
    k = c // g
    x1 = x0 * k
    y1 = y0 * k
    return f"Nghiệm riêng: (x1, y1) = ({x1}, {y1}), Nghiệm tổng quát: x = {x1} + {b//g}t, y = {y1} - {a//g}t, t ∈ Z"
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))
c = int(input("Nhập số nguyên c: "))
result = solve_diophantine(a, b, c)
print(result)
```
