> 1. Viết chương trình mã hoá và giải mã bằng mã mũ

```python
import math

def is_prime(n):
    if n <= 1: return False
    for i in range(2, int(math.isqrt(n)) + 1):
        if n % i == 0: return False
    return True

def generate_keys(p, e):
    if not is_prime(p):
        raise ValueError("p phải là một số nguyên tố.")
    
    phi = p - 1
    if math.gcd(e, phi) != 1:
        raise ValueError("e và p-1 phải nguyên tố cùng nhau.")
    
    d = pow(e, -1, phi)
    return e, d

def encrypt(m, e, p):
    if m >= p:
        raise ValueError("Thông điệp m phải nhỏ hơn p.")
    return pow(m, e, p)

def decrypt(c, d, p):
    return pow(c, d, p)

# Example
p = 61  
e = 17
m = 42
e, d = generate_keys(p, e)
c = encrypt(m, e, p)
print(f"Khoá công khai: (e={e}, p={p})")
print(f"Khoá bí mật: (d={d}, p={p})")
print(f"Thông điệp gốc: {m}")
print(f"Bản mã: {c}")
```
